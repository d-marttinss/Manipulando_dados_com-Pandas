## Criando Agrupamentos

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
dados['Bairro'].drop_duplicates()
```

```python
bairros = list (dados['Bairro'].drop_duplicates().values)
```

```python
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
## Criando a varíável grupo_bairros e criando uma agrupação com o groupby selecionando a coluna de bairro para fazer esse
## agrupamento
grupo_bairros = dados.groupby('Bairro')
```

```python
grupo_bairros.groups
```

```python
grupo_bairros[['Valor', 'Condominio']].mean().round(2)
```

## Estatísticas Descritivas

```python
grupo_bairros['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Minimo', 'max': 'Máximo'})
```

```python
## Biblioteca de análise gráfica dos dados -> Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_bairros['Valor'].mean().plot.bar(color = "blue")
fig.set_ylabel('Valor do aluguel', fontsize = (15))
fig.set_title('Valor médio de aluguel por bairro', fontsize = (20))
```

```python
fig = grupo_bairros['Valor'].max().plot.bar(color = "blue")
fig.set_ylabel('Valor do aluguel', fontsize = (15))
fig.set_title('Valor máximo de aluguel por bairro', fontsize = (20))
```
